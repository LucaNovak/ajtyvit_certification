## Dataset

| `age` | Vek zákazníka v rokoch | Vek klienta v čase zberu dát. Dôležitý faktor pre pochopenie vekových skupín a ich tendencie k odchodu |
| `tenure` | Počet rokov ako klient banky | Dĺžka vzťahu s bankou v rokoch. Indikuje lojalitu zákazníka a stabilitu vzťahu |
| `balance` | Zostatok na účte | Aktuálny zostatok na všetkých účtoch klienta v banke. Vyjadruje finančnú aktivitu klienta |
| `products_number` | Počet využívaných bankových produktov | Počet rôznych produktov, ktoré klient aktívne využíva (účty, karty, poistenia, úvery). Ukazuje mieru angažovanosti klienta |
| `credit_card` | Či má zákazník kreditnú kartu | Indikátor (1=Áno, 0=Nie) vlastníctva kreditnej karty. Dôležitý ukazovateľ využívania úverových produktov |
| `active_member` | Či je zákazník aktívnym členom | Indikátor (1=Áno, 0=Nie) aktivity klienta. Aktívny klient pravidelne využíva bankové služby |

| `churn` | Či zákazník odišiel | Cieľová premenná (1=Áno, 0=Nie) indikujúca, či klient ukončil svoje služby v banke |

## Data splits

### Predictors / target split

```python
# TODO
# Predictors / Target split
x = df.iloc[:, :-1].copy()   
y = df.iloc[:, -1].copy()   
```

### Train / test split

```python
# TODO
#Train/Test split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=999)
```

## Create model / pipeline

```python
#Preprocessing data

# Columns we can delete (not predictive)
drop_columns=['customer_id']

# Columns representing nominal categories (for OneHot encoding)
category_nominal_columns= [
    'country','gender','products_number'
]

    
#Helper function
def remove_columns(x):
    x=x.copy()
    return x.drop(columns=drop_columns)
```

```python
df.head(2)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import FunctionTransformer
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

# Helper function, that forces transformers to operate only on some columns, not whole dataframe
def per_column_transformers(transformers_list, remainder='passthrough'):
    return ColumnTransformer(transformers=transformers_list, remainder=remainder, verbose_feature_names_out=False)

# Create pipeline
pipeline = Pipeline(steps=[
    ('remove_columns', FunctionTransformer(remove_columns)), 
    ('per_column_transformers', per_column_transformers([
        ('convert_nominal_categories', OneHotEncoder(sparse_output=False, handle_unknown='ignore'), category_nominal_columns)

    ])),
])

# Configure pipeline
pipeline.set_output(transform="pandas") 

```

```python
x_train_transformed = pipeline.fit_transform(x_train)

# Check result
x_train_transformed.head(2)
```

```python
# Add model to pipeline:

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import FunctionTransformer
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from xgboost import XGBClassifier

# Helper function, that forces transformers to operate only on some columns, not whole dataframe
def per_column_transformers(transformers_list, remainder='passthrough'):
    return ColumnTransformer(transformers=transformers_list, remainder=remainder, verbose_feature_names_out=False)

# Create pipeline
pipeline = Pipeline(steps=[
    ('remove_columns', FunctionTransformer(remove_columns)), 
    ('per_column_transformers', per_column_transformers([
        ('convert_nominal_categories', OneHotEncoder(sparse_output=False, handle_unknown='ignore'), category_nominal_columns)
    ])),
    ('model', XGBClassifier(random_state=555))
])

# Configure pipeline
pipeline.set_output(transform="pandas") 

```

## Model Tuning

```python
#Nem imports
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint, uniform

# Define the parameter space
param_dist = {
    'model__n_estimators': randint(300, 601),
    'model__max_depth': randint(3, 4),
    'model__learning_rate': uniform(0.01, 0.05),
    'model__colsample_bytree': uniform(0.2, 0.4)
}

# Set up RandomizedSearchCV
random_search = RandomizedSearchCV(pipeline, param_dist, n_iter=50, cv=10, scoring='roc_auc', random_state=341, n_jobs=-1)

# Fit the random search
random_search.fit(x_train, y_train)
```

```python
random_search.best_params_
```

```python
random_search.best_score_
```

```python
best_model=random_search.best_estimator_
best_model
```

## Model Serialization

```python
import joblib 

joblib.dump(best_model, "model_xgb.pkl")
```

```python
best_model=joblib.load("model_xgb.pkl")
best_model
```

## Model Evaluation

### Evaluate on test-set

```python
# Make predictions on test-set

from sklearn.metrics import roc_auc_score

y_pred = best_model.predict(x_test)

roc_auc = roc_auc_score(y_test, y_pred)
print(roc_auc)
```
